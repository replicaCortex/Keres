```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

#    x
x = np.linspace(0, 20, 5000)

#      (PDF)  - 
for df in [1, 2, 3, 4, 5]:
	pdf = chi2.pdf(x, df)

	#  
	plt.plot(x, pdf, label=f'- (df={df})')
	plt.title('- ')
	plt.xlabel('x')
	plt.ylabel(' ')
	plt.legend()
	plt.grid(True)
	plt.show()
```

```python
df = 9
pdf = chi2.pdf(x, df)

#  
plt.plot(x, pdf, label=f'- (df={df})')
plt.title('- ')
plt.xlabel('x')
plt.ylabel(' ')
plt.legend()
plt.grid(True)
plt.show()
```
